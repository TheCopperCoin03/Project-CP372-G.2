**Feature Engineering**

```python
# สมมุติcost = 70%ของราคาขาย
df1['Cost'] = df1['Purchase Amount (USD)'] * 0.7

# คำนวณกำไร(Profit)
df1['Profit'] = df1['Purchase Amount (USD)'] - df1['Cost']

# คำนวณอัตรากำไร(Profit Margin)
df1['Profit Margin'] = df1['Profit'] / df1['Purchase Amount (USD)']
```

```python
#สร้างกลุ่มอายุ(Age Group)
df1['Age Group'] = pd.cut(df1['Age'],
                         bins=[0, 18, 30, 45, 60, 100],
                         labels=['Teen', 'Young Adult', 'Adult', 'Mid Age', 'Senior'])
```

```python
#สร้างฟีเจอร์ High Spender(ผู้ที่ใช้จ่ายมากกว่าเกณฑ์ เช่น 100 USD)
df1['High Spender'] = df1['Purchase Amount (USD)'] > 100
```

```python
#สร้างฟีเจอร์ Loyal Customer(ลูกค้าที่ซื้อซ้ำเกิน 5 ครั้ง)
df1['Loyal Customer'] = df1['Previous Purchases'] >= 5
```

```python
#แปลงฟีเจอร์แบบ Yes/No ให้เป็นตัวเลข
df1['Discount Applied'] = df1['Discount Applied'].map({'Yes': 1, 'No': 0})
df1['Promo Code Used'] = df1['Promo Code Used'].map({'Yes': 1, 'No': 0})
```

```python
#สร้างฟีเจอร์ Preferred Payment Match(ตรงกับวิธีจ่ายจริงหรือไม่)
df1['Preferred Payment Match'] = (df1['Payment Method'] == df1['Preferred Payment Method']).astype(int)
```

## Data Cleaning

- High Spender = ผู้ที่มียอดใช้จ่าย > 100 USD
- Loyal Customer = ลูกค้าที่มีการซื้อซ้ำ ≥ 5 ครั้ง

- Promo Code Used = แปลงเป็น 1 (Yes), 0 (No)
- Preferred Payment Match = วิธีจ่ายเงินจริงตรงกับ Preferred Method หรือไม่ (1=ตรง, 0=ไม่ตรง)
